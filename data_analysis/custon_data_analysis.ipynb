```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Analysis of my Customized Dataset 

```python
# load the data
df = pd.read_csv("../../my_data/identification-dataset/my_custom_data/big-identification-dataset.csv")
```

Display basic information of the dataset 

```python
df.dropna(inplace=True)
df.info()
```

## Investigation of the Class(NLOS material)-Distribution

```python
# Get the material names
material_names = df.material.unique()
material_names
```

```python
# Insert a fist value for LOS condition
material_names = np.insert(material_names, 0, 'nothing')
material_names
```

```python
# Calculate value counts for the categorical column
value_counts = df.NLOS_material.value_counts()

# Create a bar plot
plt.bar(value_counts.index, value_counts.values)

# Add labels and title
plt.xlabel('NLOS Material')
plt.ylabel('Frequency')
plt.title('Distribution of NLOS Material')

# Rename the x-axis ticks
plt.xticks(rotation=45)  # Rotates the tick labels for better visibility
plt.xticks(np.arange(len(material_names)), material_names)  # Assign the tick labels from the array


# Show the plot
plt.show()
```
